# Gradient Descent as a Fixed-Point Method

## 1. Introduction
Gradient Descent is an iterative optimization algorithm used to find minima of functions. It can be understood as a **fixed-point iteration method**, where the update rule moves step-by-step towards a point where the gradient is zero (a critical point).


A **fixed-point iteration** is any iterative process of the form:
$$x_{k+1} = g(x_k)$$
where $ g(x)$ is an iteration function. The goal is to find a fixed point $ x^*$ such that $ g(x^*) = x^*$.

In **gradient descent**, we update $ x_k $ using the rule:
$$
x_{k+1} = x_k - \alpha \nabla f(x_k),
$$
where:
- $ \alpha $ is the step size (learning rate),
- $ \nabla f(x_k) $ is the gradient of $ f(x) $, pointing in the direction of the steepest ascent.

If we define the iteration function as:
$$
g(x) = x - \alpha \nabla f(x),
$$
then gradient descent is a fixed-point iteration where we seek a fixed point $ x^* $ such that:
$$
x^* = g(x^*) = x^* - \alpha \nabla f(x^*).
$$
Since $ x^* $ does not change, this implies $ \nabla f(x^*) = 0 $, meaning that $ x^* $ is a critical point of $ f(x) $.


## 2. Geometric Interpretation
Geometrically, gradient descent can be visualized as follows:
- The function $f(x)$ represents a surface.
- The gradient $\nabla f(x)$ gives the direction of the steepest increase.
- By moving in the opposite direction, we descend towards a local minimum.
- The algorithm iterates until it reaches a point where $\nabla f(x) = 0$, meaning no further descent is possible.

### Example: Minimizing a Quadratic Function
Consider the simple function:
$$
f(x) = \frac{1}{2} x^2.
$$
Its gradient is:
$$
\nabla f(x) = x.
$$
The gradient descent update rule becomes:
$$
x_{k+1} = x_k - \alpha x_k = (1 - \alpha) x_k.
$$
This is a **contractive fixed-point iteration**, ensuring convergence if $0 < \alpha < 2$.


````python
## 3. Python Implementation
# Hereâ€™s how we can implement gradient descent to minimize a simple quadratic function.

# ```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 0.5 * x**2

def grad_f(x):
    return x  # Gradient of f(x) = 0.5*x^2 is just x

def gradient_descent(x0, alpha, iterations):
    x_values = [x0]
    for _ in range(iterations):
        x0 = x0 - alpha * grad_f(x0)
        x_values.append(x0)
    return x_values

# Parameters
x0 = 10  # Initial value
alpha = 0.1  # Learning rate
iterations = 10

# Run gradient descent
x_vals = gradient_descent(x0, alpha, iterations)

# Plot the function and iterations
x_range = np.linspace(-12, 12, 100)
y_range = f(x_range)
plt.plot(x_range, y_range, label='f(x) = 0.5x^2')
plt.scatter(x_vals, [f(x) for x in x_vals], color='red', label='Iterations')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent as Fixed-Point Iteration')
plt.show()
# ```

### Output:
# - The red points show the iteration steps moving towards the minimum at $ x = 0 $.
````




## 4. Example with a Non-Quadratic Function
Let's now consider a non-quadratic function:
$$
f(x) = \ln(1 + x^2).
$$
The gradient is:
$$
\nabla f(x) = \frac{2x}{1 + x^2}.
$$
We can modify our Python code to apply gradient descent on this function.


````python
# Parameters
x0 = 2 # Initial value
alpha = 0.001  # Learning rate
iterations = 100
````

````python
def f(x):
    return np.log(1 + x**2)

def grad_f(x):
    return (2 * x) / (1 + x**2)
def gradient_descent(x0, alpha, iterations):
    x_values = [x0]
    for _ in range(iterations):
        x0 = x0 - alpha * grad_f(x0)
        x_values.append(x0)
    return x_values
# Run gradient descent

x_vals = gradient_descent(x0, alpha, iterations)

# Plot
x_range = np.linspace(-5, 5, 100)
y_range = f(x_range)
plt.plot(x_range, y_range, label='f(x) = ln(1 + x^2)')
plt.scatter(x_vals, [f(x) for x in x_vals], color='red', label='Iterations')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent on a Non-Quadratic Function')
plt.show()
````

## 5. Key Takeaways
1. **Gradient Descent as Fixed-Point Iteration:**
   - The update rule can be written as $g(x) = x - \alpha \nabla f(x)$, making it a fixed-point method.
   - The fixed points satisfy $\nabla f(x^*) = 0$, meaning they are critical points of $f(x)$.
2. **Geometric Understanding:**
   - The gradient points in the direction of steepest ascent; moving opposite finds minima.
   - The update rule effectively moves the point closer to a minimum step-by-step.
3. **Examples in Python:**
   - We visualized gradient descent on quadratic and non-quadratic functions.
   - The algorithm successfully converged to the function's minimum.

Gradient descent is a fundamental concept in optimization and machine learning, and understanding it as a **fixed-point method** provides deeper insights into its behavior and convergence properties.
