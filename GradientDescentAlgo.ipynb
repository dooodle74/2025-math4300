# Gradient Descent

```python
# Gradient descent for linear regression
# y = actual value of function/data
# y_pred = wx + b predicted value (some weight w and bias b)
import numpy as np
```

```python
x = np.random.randn(10, 1)
y = 3*x + np.random.rand()
#Parameters
w = 0.0
b = 0.0
# Hyper Parameter
learning_rate = 0.1

```

```python
def descend(x,y,w,b,learning_rate):
    # l = (y - (wx + b))**2 Minimize loss function 
    dldw = 0.0
    dldb = 0.0
    N = x.shape[0]
    for xi,yi in zip(x,y):
        dldw += -2*xi*(yi-(w*xi+b))
        dldb += -2*(yi-(w*xi+b))
    w = w - learning_rate*(1/N)*dldw
    b = b - learning_rate*(1/N)*dldb
    return w,b 
```

```python
for epoch in range(400):
    w,b=descend(x,y,w,b,learning_rate)
    y_pred = w*x+b
    loss = np.divide(np.sum((y-y_pred)**2,axis=0),x.shape[0])
    print(f'{epoch} loss is {loss}, parameters w:{w}, b:{b}')
```
